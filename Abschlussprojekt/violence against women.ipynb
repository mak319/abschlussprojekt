```python
import pandas as pd

data_df = pd.read_csv("Domestic violence.csv")  # Load the CSV file
display(data_df.head(100))
```

```python
import numpy as np

# Leere Spalte mit "Keine" als Standardwert
data_df["Type_of_Violence"] = "None"

# Nur bei 'yes' in der Violence-Spalte fÃ¼llen wir mit zufÃ¤lligen Arten
arten_von_gewalt = ["Physical", "Mentally abuse", "Sexual Abuse"]

# Indexe, wo Gewalt gemeldet wurde
yes_index = data_df[data_df['Violence '] == "yes"].index

# ZufÃ¤llige Gewaltarten generieren
np.random.seed(42)  # fÃ¼r reproduzierbare Ergebnisse
zufaellige_gewaltarten = np.random.choice(arten_von_gewalt, size=len(yes_index))

# Eintragen
data_df.loc[yes_index, "Type_of_Violence"] = zufaellige_gewaltarten

data_df
```

```python
# Gruppierung zwischen Bildungsniveau und hÃ¤uslicher Gewalt , habe ich die ProzentsÃ¤tze von Gewalt- und Nicht-GewaltfÃ¤llen nach Bildungsniveau berechnet

query = data_df.groupby('Education ')['Violence '].value_counts()
yes = query.reset_index(name="Count")
yes = yes.rename(columns={"Count": "Violence Count"})  # Correcting this line
yes['Total Violence per Education'] = yes.groupby('Education ')['Violence Count'].transform('sum')
yes['Percentage by Education'] = (yes['Violence Count'] * 100 / yes['Total Violence per Education']).round(2)
display(yes)
```

```python
#Prozentangaben von verheiratetn und unverheirateten von
status=data_df.groupby('Marital status ')['Violence '].value_counts().reset_index()
status = status[status['Violence '] == 'yes']
total = status.loc[1:,'count'].sum()
status.loc[1, 'Percentage'] = (status.iloc[0,2]*100/total).round(2)
status.loc[3, 'Percentage'] = (status.iloc[1,2]*100/total).round(2)
display(status)
```

```python
# Die Analyse zeigt die Gewaltanteile bei den verschiedenen BeschÃ¤ftigungsstatus Ã¤hnlich sind, 
# insbesondere zwischen den ErwerbstÃ¤tigen und den Arbeitslosen.
employment = data_df.groupby('Employment ')['Violence '].value_counts().reset_index()
employment = employment.rename(columns={'count': 'Violence Count'})
employment['Total'] = employment.groupby('Employment ')['Violence Count'].transform('sum')
employment['Percentage'] = (employment['Violence Count'] * 100 / employment['Total']).round(2)
display(employment)
```
